To test true intelligence, they split the data into three levels of difficulty:
1. Cross-Task: The agent sees a known website but has to perform a new task.
2. Cross-Website: The agent has to perform tasks on a website it has never seen.
3. Cross-Domain: The agent acts on a completely new category of website (e.g., trained on shopping, tested on flight booking).

```python
from datasets import load_dataset

SLURM_PATH = "/home/yandex/MLWG2025/amitr5"
CACHE_DIR = f"{SLURM_PATH}/tmp/hf_cache"
def count_annotations(split):
    return len(set(split["annotation_id"]))


ds = load_dataset("osunlp/Multimodal-Mind2Web", cache_dir=CACHE_DIR)

print("Dataset splits:", list(ds.keys()))
test_domain_ds = ds["test_domain"]
test_task_ds = ds["test_task"]
test_website_ds = ds["test_website"]



print(f"Number of samples in test_domain:  {len(test_domain_ds)} (annotations: {count_annotations(test_domain_ds)})")
print(f"Number of samples in test_task:    {len(test_task_ds)} (annotations: {count_annotations(test_task_ds)})")
print(f"Number of samples in test_website: {len(test_website_ds)} (annotations: {count_annotations(test_website_ds)})")
print("Total number of test samples:", len(test_domain_ds) + len(test_task_ds) + len(test_website_ds))
print("Total number of test annotations:", count_annotations(test_domain_ds) + count_annotations(test_task_ds) + count_annotations(test_website_ds))
```
